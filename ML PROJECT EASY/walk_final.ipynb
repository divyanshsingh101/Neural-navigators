```python
import tensorflow as tf
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('WalkTheDogs.csv')
dataset.head()
#in walk column of dataset :0 signifies light walk and 1 signifies run
```

```python
dataset.drop(['Day','Steps'],axis=1,inplace=True)
dataset
```

```python
Weather=pd.get_dummies(dataset['Weather'])
dataset['StepCount']=dataset['StepCount']/1000
```

```python
df=pd.concat([dataset,Weather],axis=1)
df.drop('Weather',axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
y=df['Kcal']
df.drop('Kcal',axis=1,inplace=True)
x=df.iloc[:,1:7]
x
```

```python
X_train,X_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=42)
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train
```

```python
from tensorflow.keras import layers
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import  LeakyReLU,PReLU,ELU,ReLU
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
classifier=Sequential()
classifier.add(Dense(units=100,activation='relu', kernel_initializer='he_normal',input_shape=(6,)))
classifier.add(Dense(units=80,kernel_initializer='he_normal',activation='relu'))
classifier.add(Dense(units=45,kernel_initializer='he_normal',activation='relu'))
classifier.add(Dense(1,kernel_initializer='he_normal',activation='linear'))
classifier.compile(optimizer='adam',loss='mean_squared_error',metrics=['mse'])
```

```python
classifier.fit(X_train, y_train, epochs=150,validation_split=0.33, batch_size=32)
```
