```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#REM Refers REM stage of sleep in percentage 
#similarly DSP,LSP Also refer to the stages of sleep in percentage
# Sd === Sleeep Duration in hours
# Se === Sleeep Eficiency
```

```python
df = pd.read_csv('/Users\Divyansh\Desktop\ML PROJECT EASY\Sleep_Efficiency.csv')
df.head()
```

```python
df['Sd'] = df['Sd'] / 24
df['REM'] = df['REM'] / 100
df['Dsp'] = df['Dsp'] / 100
df['Lsp'] = df['Lsp'] / 100
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Sd', 'REM', 'Dsp', 'Lsp']], df['Se'], test_size=0.15, random_state=0)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = Sequential([
layers.Dense(64, input_dim=4, activation='relu'),  
layers.Dense(32, activation='relu'),              
layers.Dense(1, activation='linear'),  
])
    
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(X_train_scaled, y_train, epochs=100, batch_size=32)
```

```python
import gradio as gd
import numpy as np

# Assuming 'model' and 'scaler' are defined and loaded elsewhere in your code

# Define the function for prediction and tips
def predict_se_nn(Sd, REM, Dsp, Lsp):
    input_features = np.array([[Sd/24, REM/100, Dsp/100, Lsp/100]])
    input_features_scaled = scaler.transform(input_features) 
    predicted_se = ann.predict(input_features_scaled)
    predicted_se_percentage = predicted_se[0][0] * 100
    
    # Provide tips based on the predicted sleep efficiency
    if predicted_se_percentage >= 90:
        tips = "Your predicted sleep efficiency is excellent! Congratulations! Here are some tips to maintain it:"
        tips += "\n- Stick to a consistent sleep schedule, even on weekends."
        tips += "\n- Create a relaxing bedtime routine to signal your body it's time to sleep."
        tips += "\n- Ensure your sleep environment is comfortable, quiet, and dark."
        tips += "\n- Limit exposure to screens (phones, computers, TVs) before bedtime."
    elif 80 <= predicted_se_percentage < 90:
        tips = "Your predicted sleep efficiency is good. Here are some tips to maintain it:"
        tips += "\n- Keep a consistent sleep schedule, aiming for at least 7-9 hours of sleep per night."
        tips += "\n- Establish a relaxing bedtime routine to help you wind down before sleep."
        tips += "\n- Make sure your bedroom environment is conducive to sleep."
    elif 70 <= predicted_se_percentage < 80:
        tips = "Your predicted sleep efficiency is fair. Here are some tips to improve it:"
        tips += "\n- Try to stick to a regular sleep schedule, even on weekends."
        tips += "\n- Create a calming bedtime routine to help signal your body that it's time to sleep."
        tips += "\n- Consider addressing any sources of noise or light that may disrupt your sleep."
        tips += "\n- Limit caffeine and alcohol intake, especially in the hours leading up to bedtime."
    elif 60 <= predicted_se_percentage < 70:
        tips = "Your predicted sleep efficiency needs improvement. Here are some tips to enhance it:"
        tips += "\n- Aim for a consistent sleep schedule, going to bed and waking up at the same time every day."
        tips += "\n- Establish a relaxing bedtime routine, such as reading or taking a warm bath."
        tips += "\n- Evaluate your sleep environment for factors that may be hindering your sleep, such as noise or light."
        tips += "\n- Limit exposure to screens before bedtime, as the blue light emitted can interfere with sleep."
        tips += "\n- Consider incorporating relaxation techniques, such as deep breathing or meditation, to help you fall asleep."
    else:
        tips = "Your predicted sleep efficiency is poor. It's essential to prioritize improving your sleep. Here are some recommendations:"
        tips += "\n- Focus on creating a consistent sleep schedule, aiming for at least 7-9 hours of sleep each night."
        tips += "\n- Develop a relaxing bedtime routine to help prepare your body and mind for sleep."
        tips += "\n- Make sure your sleep environment is conducive to sleep, addressing any sources of noise, light, or discomfort."
        tips += "\n- Limit caffeine and alcohol intake, particularly in the hours leading up to bedtime."
        tips += "\n- Consider seeking advice from a healthcare professional if you continue to experience poor sleep despite making lifestyle changes."
    
    return f"Predicted Sleep Efficiency: {predicted_se_percentage:.2f}%\n\n{tips}"

# Define the Gradio interface
headline = "SLEEP EFFICIENCY PREDICTION"
description = "This tool predicts your sleep efficiency based on input sleep metrics and provides tips."
inputs = [
    gd.Number(label="Sleep Duration (hours)"),
    gd.Number(label="REM Percentage (%)"),
    gd.Number(label="Deep Sleep Percentage (%)"),
    gd.Number(label="Light Sleep Percentage (%)")
]

output = gd.Textbox(label="Predicted Sleep Efficiency and Tips")

interface = gd.Interface(
    fn=predict_se_nn,
    inputs=inputs,
    outputs=output,
    title=headline,
    description=description
)

# Launch the Gradio interface
interface.launch()
```
